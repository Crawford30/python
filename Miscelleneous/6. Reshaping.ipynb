```python
# Import the libraries required
import numpy as np
import pandas as pd
```

```python
# Simulating a sales dataset

year = [2014,2015,2016,2017,2018]
Q1 = np.random.choice(range(60,100,1),5,replace= False, p=None)
Q2 = np.random.choice(range(60,100,1),5,replace= False, p=None)
Q3 = np.random.choice(range(60,100,1),5,replace= False, p=None)
Q4 = np.random.choice(range(60,100,1),5,replace= False, p=None)

sales = pd.DataFrame({"Year":year,"Q1":Q1,"Q2":Q2,"Q3":Q3,"Q4":Q4})
sales
```

```python
# Reshape the sales dataset from wide to long
saleslong = pd.melt(sales,
                    id_vars = "Year",
                    value_vars = ["Q1","Q2","Q3","Q4"],
                    var_name = "Quarters",
                    value_name ="Sales")
saleslong
```

```python
# Reshape the sales dataset from long to wide
saleswide = pd.DataFrame(pd.pivot(saleslong.Year,saleslong.Quarters,saleslong.Sales))
saleswide["Year"] = saleswide.index
saleswide
```
