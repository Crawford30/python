```python
## Import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
## Simulate a dataset
gender=np.random.choice(["Male","Female"],100,replace=True)
treatment = np.random.choice(["Treatment","Control"],100,replace=True)
towns = np.random.choice(["Kinoo","Kawangware","Nyeri","Kisumu"],100,replace=True)
idd = list(range(1,101,1))
income = np.random.choice(list(range(10000,100000,785)),100,replace=True)
dataset = pd.DataFrame({"Users":idd,"Gender":gender,"Treatment":treatment,"Income":income,"Town":towns})
#dataset["Users2"] = print("Person "+ dataset["Users"])
#del dataset["Users"]
#dataset.columns = ["Gender","Treatment","Users2"]
dataset
```

```python
## Plot well labelled bar graphs of treatment and gender
treatment_count = pd.DataFrame(dataset["Treatment"].value_counts())
treatment_count
```

```python
plt.barh(treatment_count.index,treatment_count.Treatment,color = "blue")

plt.xlabel("Treatment Condition")
plt.ylabel("Treatment Frequencies")
```
