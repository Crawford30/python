```python
## Create a table that counts the number of observations in each category, of each categorical variable
#popsize = pd.DataFrame(townsdataset.groupby(["Party"])["PopSize"].mean().reset_index())

# catcount = pd.DataFrame(bankdata.groupby(["marital"])["age"].count().reset_index())
# catcount
def catcount (var):
    catcountb = pd.DataFrame(bankdata.groupby(var)["age"].count().reset_index())
    return(catcountb)
# catcount(bankdata.marital)
# catcount(bankdata.job)
# catcount(bankdata.education)
# catcount(bankdata.default)
# catcount(bankdata.housing)
# catcount(bankdata.loan)
# catcount(bankdata.y)

cats = ["marital","job","education","default","housing","loan","y"]
for item in cats:
    print(catcount(bankdata[item]))
```

```python
## Drawing a frequency plot for marital categories
plt.figure(figsize = (8,8))
ax = sns.countplot(x = "marital",data = bankdata, palette = "Blues_d")
ax.set_title("Distribution of Marital Categories")
ax.set_xlabel("Marital Categories")
ax.set_ylabel("Frequencies")
ax.set_xticklabels(ax.get_xticklabels(), rotation = 30)

```

```python
## Drawing a frequency plot for marital categories, by defaultment of the loan.
sns.set(font='sans-serif', font_scale=1.5,style = "white")
plt.figure(figsize = (8,8))
ax = sns.countplot(y = "marital",hue = "loan",data = bankdata, palette = "Greens_r")
ax.set(title = "A graph of Marital Status \n by \n Loan Access",ylabel="Frequencies",xlabel = "Marital Status")
```

```python
## Drawing a scatter plot of balance against duration.
sns.set(font='sans-serif', font_scale=1.5,style = "white")
plt.figure(figsize = (8,8))
#ax = sns.countplot(y = "marital",hue = "loan",data = bankdata, palette = "Greens_r")
ax = sns.pointplot( y ="balance", x = "duration",data = bankdata, palette = "Reds")
ax.set(title = "A graph of Marital Status \n by \n Loan Access",ylabel="Balance",xlabel = "Duration")

```

```python
# Boxplot of bank balance by job
plt.figure(figsize=(8,8))
ax = sns.boxplot(x = "education",y = "age",data = bankdata, color = "Brown")
ax.set(title = "A graph of Distribution of Age \n against \n Education Categories",
       xlabel = "Education Categories",ylabel = "Age" )
ax.set_xticklabels(ax.get_xticklabels(), rotation = 30)

```
