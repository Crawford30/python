```python
#3. Generate variables that will constitute the wafanyakazi dataset
subject_id = np.random.choice(list(range(10000,20000,1)),size=500,replace=False)
gender = np.random.choice(list(["Male","Female"]),size = 500,replace=True)
age = np.random.choice(list(range(18,51,1)),size = 500,replace=True)
department = np.random.choice(["Data","Operations","Finance","Research Analyst","Associate"],size = 500,replace=True)
role = np.random.choice(["Junior","Mid","Senior"],size = 500, replace = True)
income = np.random.choice(list(range(1000,10000,217)),size = 500, replace=True)
marital_status = np.random.choice(["Single","Married","Divorced"],size = 500, replace = True)
county = np.random.choice(["Nyeri","Kisumu","Taita","Embu","Kirinyaga","Laikipia","Wajir","Mombasa","Lamu",
                           "Kiambu","Nairobi"],size = 500, replace=True)
leave_days = np.random.choice(list(range(0,25,1)),size = 500, replace=True)
promotion = np.random.choice(["Yes","No"],size = 500, replace=True)
```

```python
#4. Generate the wafanyikazi dataset and save it externally

wafanyikazi =pd.DataFrame({"Sid":subject_id,"Gender":gender,"Age":age,"Department":department,
                           "Role": role,"Income": income,"Marital_Status":marital_status,"County":county,
                           "Leave_Days": leave_days,"Promotion": promotion})
## Inspecting the first 10 rows of the dataset
wafanyikazi.head(n = 10)
```

```python
#5. Outline the number of rows and columns in the dataset
wafanyikazi.shape

print("The wafanyikazi dataset has %d rows and %d columns" %(wafanyikazi.shape[0],wafanyikazi.shape[1]))
```

```python
#6. Inspect the data types of the variables in the dataset
wafanyikazi.dtypes
```

```python
#7. Convert the non-numerical variables to categorical. 

for col in ["Gender","Department","Role","Marital_Status","County","Promotion"]:
    wafanyikazi[col] = wafanyikazi[col].astype('category')
## inspect if the loop has worked.
wafanyikazi.dtypes    
```

```python
#8. Generate tables that display the number of entries per category, for each categorical variable

for col in ["Gender","Department","Role","Marital_Status","County","Promotion"]:
    print(pd.DataFrame(wafanyikazi[col].value_counts()))
    
# counter = pd.DataFrame(wafanyikazi.Gender.value_counts())
# counter
```

```python
#9. Generate tables that display various summary statistics, for each continuous variable
for col in ["Age","Income","Leave_Days"]:
    print(pd.DataFrame(wafanyikazi[col].describe()))
```

```python
#10. Generate count graphs of all categorical variables
# plt.figure(figsize = (5,5))
# sns.set(font='sans-serif', font_scale=1.5,style = "white")
# ax = sns.countplot(x = "Gender",data = wafanyikazi, palette = "Blues")
# ax.set(title = "Distribution of Gender",xlabel = "Gender",ylabel = "Frequency")
# ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
```

```python
for var in ["Gender","Department","Role","Marital_Status","County","Promotion"]:
    plt.figure(figsize = (8,8))
    sns.set(font='sans-serif', font_scale=1.5,style = "white")
    ax = sns.countplot(x = var,data = wafanyikazi, palette = "Blues")
    ax.set(title = "Distribution of "+ var,xlabel = var,ylabel = "Frequency")
    ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
    print(ax)
```

```python
#11.Plots showing the distribution of Role, Department, Marital Status and Promotion by Gender
for var in ["Role","Department","Marital_Status","Promotion"]:
    plt.figure(figsize = (8,8))
    ax = sns.countplot(x = var, hue = "Gender",data=wafanyikazi,palette = "Blues")
    ax.set(title = ("Distribution of "+var+"\n by \n Gender"),xlabel = var,ylabel = "Frequency",ylim = (0,160))
    ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
    print(ax)
```

```python
#12. Boxplots of the continuous variables, by the various categorical variables

# plt.figure(figsize = (8,8))
# ax = sns.boxplot(x = "Gender",y = "Income",data = wafanyikazi, palette = "Blues")
# ax.set(title = "Distribution of Income \n by \n Gender",xlabel = "Gender",ylabel = "Income")

for conts in ["Income","Age","Leave_Days"]:
    for categos in ["Gender","Department","Role","Marital_Status","Promotion"]:
        plt.figure(figsize = (8,8))
        ax = sns.boxplot(x = categos, y = conts, data = wafanyikazi,palette = "Reds" )
        ax.set(title  = "Distribution of "+conts+"\n by \n" + categos,xlabel = categos, ylabel = conts)
        ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
```

```python
#13. Plotting the average income per each categorical variable

for col in ["Gender","Promotion","Role","Marital_Status","Department"]:
    mean_income = pd.DataFrame(wafanyikazi.groupby(col)["Income"].mean().reset_index())
    print(mean_income)
    plt.figure(figsize = (8,8))
    ax = sns.barplot(x = col, y = "Income",data = mean_income, palette = "Greens")
    ax.set(title = "Average Income Amount \n by \n"+ col,xlabel = col,ylabel = "Average Income Amount")
    ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
    
```

```python
#14. Plotting the average 'cont variable' per each categorical variable
for col in ["Gender","Promotion","Role","Marital_Status","Department"]:
    for col2 in ["Income","Age","Leave_Days"]:
        mean_var = pd.DataFrame(wafanyikazi.groupby(col)[col2].mean().reset_index())
        print(mean_var)
        plt.figure(figsize = (8,8))
        ax = sns.barplot(x = col, y = col2,data = mean_var, palette = "Greens")
        ax.set(title = "Average " +col2+ " \n by \n"+ col,xlabel = col,ylabel = "Average "+col2)
        ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
    
```
